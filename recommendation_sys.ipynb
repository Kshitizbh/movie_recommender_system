```python
#storing important column into df
df = df[['movie_id', 'title','overview','genres','keywords','cast','crew']]
```

```python
#removing null value from all row
df.dropna(inplace = True)
```

```python
#checking the duplicate value 
df.duplicated().sum()
```

```python
#importing abstract syntax tree(ast)
import ast
```

```python
#literal basically means the value, ast_literal_eval evaluates sting containing python values
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#applying convert function to keywords
df.keywords = df.keywords.apply(convert)
```

```python
#applying convert function to genres
df.genres = df.genres.apply(convert)
```

```python
df
```

```python
# extracting the value of key "name", only 3
def convert_(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
#applying convert_ function
df.cast = df.cast.apply(convert_)
```

```python
#extracting the value of key " job" where the value is "director" and appending the name in L
def fetch (s):
    L = []
    for i in ast.literal_eval(s):
        if i ['job'] == 'Director':
            L.append(i['name'])
    return L
        
```

```python
df.crew = df.crew.apply(fetch)
```

```python
df.overview = df.overview.apply(lambda x:x.split())
```

```python
#replacing the space between words
df.genres = df.genres.apply(lambda x:[i.replace(" ","")for i in x])
df.keywords = df.keywords.apply(lambda x:[i.replace(" ","")for i in x])
df.cast = df.cast.apply(lambda x:[i.replace(" ","")for i in x])
df.crew = df.crew.apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df
```

```python
#combining the features into one column
df['tags'] = df.overview + df.genres+ df.cast+df.crew
```

```python
df.head()
```

```python
#droping other unnecessary column
df.drop(['overview','genres','keywords','cast','crew'], axis = 1, inplace = True)
```

#### feature extraction.
### converting text into feature vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
len(cv.get_feature_names())
```

```python
import nltk
```

### ***stemming***

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(s):
    y=[]
    for i in s.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
df.tags = df.tags.apply(stem)
```

```python
df
```

### ***cosine similarity***
### ***getting similarity score using cosine_similarity***

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6]
```

```python
def recommend(s):
    movie_index = df[df['title'] == s].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```

```python
recommend('Iron Man 3')
```
